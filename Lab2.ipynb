```python
# 12. Write a function to find min and max values of any column

def find_min_max(df,col_name):
    if col_name in df.columns:
     min_value = df[col_name].min()
     max_value = df[col_name].max()
     return min_value,max_value
    else:
     return f"Column '{column_name}' not found in dataset"

Horsepower = find_min_max(df,"Horsepower")

print(Horsepower)
```

```python
# 13. Call the above function to find min max of horsepower, length, fuel efficiency
horsepower  = find_min_max(df,"Horsepower")
length   = find_min_max(df,"Length")
fuel_efficiency  = find_min_max(df,"Fuel efficiency")

print(horsepower, length, fuel_efficiency)
```
